```python
#Checking for corrolation
corrolation_matrix=df.corr(method ='pearson')

sns.heatmap(corrolation_matrix,annot=True)
plt.title('Corrolation Outcome')
plt.xlabel('Features')
plt.ylabel('Features')
plt.show()
```

```python
# 2. Visualize Distributions
sns.set(style="whitegrid")
fig, axs = plt.subplots(3, 3, figsize=(15, 15))

features = df.columns[:-1]  # Exclude the Outcome column
for i, feature in enumerate(features):
    sns.histplot(df[feature], kde=True, ax=axs[i//3, i%3], color='skyblue')
    axs[i//3, i%3].set_title(f'Distribution of {feature}')

plt.tight_layout()
plt.show()
```
